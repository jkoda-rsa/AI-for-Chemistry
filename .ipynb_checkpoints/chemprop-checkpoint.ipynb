```python

from sklearn.metrics import mean_squared_error
from math import sqrt

true_data = pd.read_csv("data/HTL_table_clean.csv")

pred_data = pd.read_csv("/Users/anthojen/Documents/EPFL_courses/model_chemprop_first/prediction/prediction.csv")

true_dict = dict(zip(true_data['SMILES'], true_data['PCE (%)']))
pred_dict = dict(zip(pred_data['SMILES'], pred_data['PCE (%)']))

common_smiles = [smiles for smiles in true_dict if smiles in pred_dict]

true_pce = [true_dict[smiles] for smiles in common_smiles]
pred_pce = [pred_dict[smiles] for smiles in common_smiles]

# Calculate the RMSE
rmse = sqrt(mean_squared_error(true_pce, pred_pce))

print(f"Root Mean Squared Error: {rmse}")
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(true_pce, pred_pce)
print(f"R-squared: {r2}")
```
